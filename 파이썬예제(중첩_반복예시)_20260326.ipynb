```python
리스트_변수 = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
```

```python
print(리스트_변수[0])
print(리스트_변수[1])
print(리스트_변수[2])
```

```python
print(리스트_변수[0][0], end='\t'); print(리스트_변수[0][1], end='\t'); print(리스트_변수[0][2])
print(리스트_변수[1][0], end='\t'); print(리스트_변수[1][1], end='\t'); print(리스트_변수[1][2])
print(리스트_변수[2][0], end='\t'); print(리스트_변수[2][1], end='\t'); print(리스트_변수[2][2])
```

```python
for i in range(len(리스트_변수)):
    print(리스트_변수[i])
```

```python
for i in range(len(리스트_변수)):
    for j in range(len(리스트_변수[i])):
        print(리스트_변수[i][j], end='\t')
    print()
```

```python
for i in range(len(리스트_변수)):
    아이템_리스트_변수 = 리스트_변수[i]
    for j in range(len(아이템_리스트_변수)):
        print(아이템_리스트_변수[j], end='\t')
    print()
```

```python
def 예제_리턴():
    return 3, 6, 9

print(예제_리턴())
type(예제_리턴())
```

```python
def 예제_리턴():
    return 3, 6, 9

a = 예제_리턴()
print(type(a))
print(a)
```

```python
def 예제_리턴():
    return 3, 6, 9

a, b, c = 예제_리턴() # a, b, c = 1, 2, 3
print(type(a), type(b), type(c))
print(a, b, c)
```

```python
def 예제_리턴():
    return 3, 6, 9

a, _, _ = 예제_리턴() # a, b, c = 1, 2, 3
print(type(a), type(_), type(_))
print(a, _, _)
```
